```python
with open('students_data.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)  # Write the header row
    writer.writerows(data)  # Write the data rows

print("students_data.csv created successfully!")
```

```python
import matplotlib.pyplot as plt

# Data for plotting
locations = [row[4] for row in data[1:]]
genders = [row[3] for row in data[1:]]
english_access = [row[5] for row in data[1:]]
# Count by location
location_counts = {"Town": locations.count("Town"), "Village": locations.count("Village")}

# Count by gender
gender_counts = {"Male": genders.count("Male"), "Female": genders.count("Female")}

# Count by English education access
english_access_counts = {"Yes": english_access.count("Yes"), "No": english_access.count("No")}

# Plotting location distribution using pie chart
plt.figure(figsize=(8, 5))
plt.pie(location_counts.values(), labels=location_counts.keys(), autopct='%1.1f%%', colors=['blue', 'green'])
plt.title('Distribution by Location')
plt.show()

# Plotting gender distribution using pie chart
plt.figure(figsize=(8, 5))
plt.pie(gender_counts.values(), labels=gender_counts.keys(), autopct='%1.1f%%', colors=['purple', 'pink'])
plt.title('Gender Distribution')
plt.show()

# Plotting access to English education using pie chart
plt.figure(figsize=(8, 5))
plt.pie(english_access_counts.values(), labels=english_access_counts.keys(), autopct='%1.1f%%', colors=['orange', 'red'])
plt.title('Access to English Education')
plt.show()

# Plotting location distribution
plt.figure(figsize=(8, 5))
plt.bar(location_counts.keys(), location_counts.values(), color=['blue', 'green'])
plt.title('Distribution by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.show()

# Plotting gender distribution
plt.figure(figsize=(8, 5))
plt.bar(gender_counts.keys(), gender_counts.values(), color=['purple', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

# Plotting access to English education
plt.figure(figsize=(8, 5))
plt.bar(english_access_counts.keys(), english_access_counts.values(), color=['orange', 'red'])
plt.title('Access to English Education')
plt.xlabel('Access')
plt.ylabel('Count')
plt.show()

# Combining all information into a single pie chart
labels = ["Town", "Village", "Male", "Female", "English Access", "No English Access"]
sizes = [
    location_counts["Town"],
    location_counts["Village"],
    gender_counts["Male"],
    gender_counts["Female"],
    english_access_counts["Yes"],
    english_access_counts["No"],
]
colors = ["blue", "green", "purple", "pink", "orange", "red"]

plt.figure(figsize=(10, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors)
plt.title('Combined Distribution of Location, Gender, and English Education Access')
plt.show()

```
