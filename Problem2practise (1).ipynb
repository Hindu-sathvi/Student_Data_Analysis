```python
# Assuming the uploaded file is named 'Student2.zip'
import io
import zipfile

# Read the zip file into a BytesIO object
zip_file = zipfile.ZipFile(io.BytesIO(uploaded['Student2.zip']))

# Assuming the CSV file is named 'Student_performance_data _.csv' within the zip file
with zip_file.open('Student_performance_data _.csv') as csv_file:
    df2 = pd.read_csv(csv_file)

# Display the DataFrame
print(df2)
```
